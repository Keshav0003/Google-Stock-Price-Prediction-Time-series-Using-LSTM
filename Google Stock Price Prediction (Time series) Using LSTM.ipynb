# Read Data 

```python
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
A=pd.read_csv("K:/DATA SCIENCE/Deep Learning/Google_Stock_Price_Train.csv")
A.tail(20)
```

```python
A.isna().sum()
```

```python
trd=A[['Open']]
trd.head(20)
```

# Changed Training Data into Array

```python
import numpy as np
trd=np.array(trd)
trd[1:50]
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms.fit_transform(trd)
```

# Create Data For Training

```python
trd.shape
```

```python
xtrain=[]
ytrain=[]
for i in range(50,1258,1):
    xtrain.append(trd[i-50:i])
    ytrain.append(trd[i:i+1])
```

```python
print(xtrain[2],"\n") 
print(ytrain[2])
```

```python
type(xtrain)
```

```python
xtr=np.array(xtrain)   #changed in array
```

```python
xtr.shape
```

```python
ytr=np.array(ytrain)   #changed in array
```

```python
ytr.shape
```

# LSTM

```python
import keras
from keras.layers import LSTM,Dense,Dropout
from keras.models import Sequential
from keras.callbacks import EarlyStopping
```

```python
es=EarlyStopping(patience=50)
nn=Sequential()
nn.add(LSTM(50,return_sequences=True,input_shape=(50,1)))
nn.add(LSTM(100,return_sequences=True))
nn.add(Dropout(0.2))
nn.add(LSTM(100,return_sequences=True))
nn.add(Dropout(0.1))
nn.add(Dense(1))
```

```python
nn.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mean_absolute_error'])
nn.fit(xtr,ytr,epochs=400,validation_split=0.2,callbacks=es)
```
